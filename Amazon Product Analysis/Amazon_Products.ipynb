```python
df = df.drop(['product_url','product_photo','product_num_offers','is_best_seller','is_amazon_choice','is_prime','climate_pledge_friendly','delivery',
        'coupon_text','unit_price','unit_count'], axis=1)
```

```python
df.columns = [col.capitalize() for col in df.columns]
```

```python
df
```

```python
df['Product_original_price']=df['Product_original_price'].fillna('$0.00')
```

```python
df['Product_price']=df['Product_price'].str.replace('$', '')
df['Product_original_price']=df['Product_original_price'].str.replace('$', '')
df['Product_minimum_offer_price']=df['Product_minimum_offer_price'].str.replace('$', '')
```

```python
df
```

```python
df['Product_star_rating']=df['Product_star_rating'].fillna(0)
```

```python
df
```

```python
df['Sales_volume']=df['Sales_volume'].fillna(0)
```

```python
df
```

```python
df['Numeric_sales_volume'] = df['Sales_volume'].str.extract(r'(\d+K?)\+?')
```
